```python
def logical_xor(str1, str2):
    return bool(str1) ^ bool(str2)

B=[]
semilla = int(input("Ingrese el tamao de la semilla: "))
for j in range(semilla):
  valor=int(input("Ingrese el valor booleano 0 o 1 = "));
  B.append(valor)
print(B)

r=int(input("Ingrese r: "))
q=int(input("Ingrese q: "))
inter= (2**q)-1
empieza= int(len(B)-1)
print("It. \t Bi ")
for i in range(inter):
  if i> empieza:
     resu1=B[i-r] 
     resu2=B[i-q]
     resulog=logical_xor(resu1,resu2)
     if resulog == False:
       resulog=0
       B.append(resulog)
     else:
       resulog=1
       B.append(resulog)  
  print(str(i) + "\t"+ str(B[i]))
print(B)
```
