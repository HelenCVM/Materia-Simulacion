## Ejercicio #1

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from matplotlib import cm
from matplotlib import colors

#cargar los datos
df = pd.read_csv('/content/fifa_datos.csv')
df
```

```python
#Visualizar la columna a trabajar
df['Weight']
```

```python
#Quitar los caracteres de letras
df['Weight'] = df['Weight'].str.replace(r"[a-zA-Z]",'')
df['Weight']
```

```python
df
```

```python
df['Weight'] = df['Weight'].astype(float, errors = 'raise')

print(df.info())
```

```python
peso1 = 0
peso2 = 0
peso3 = 0
peso4 = 0
pesos=[]
for x in df['Weight']:
    if x < 125:
        peso1=peso1+1  
    elif x < 150:
        peso2=peso2+1     
    elif x < 175:
        peso3=peso3+1  
    else:
        peso4=peso4+1  

pesos.append(peso1)
pesos.append(peso2)
pesos.append(peso3)
pesos.append(peso4)
print('Lista de pesos')
pesos
```

```python
rangos = ["<125 lbs","125 - 150 lbs","150 - 175 lbs",">175 lbs"]
colores = ["#EE6055","#60D394","#AAF683","#FFD97D"]

desfase = (0, 0, 0, 0.1)
plt.pie(pesos, labels=rangos, autopct="%0.1f %%", colors=colores, explode=desfase)
plt.axis("equal")
plt.show()
```

```python
cali1=0
cali2=0
cali3=0
cali4=0
cali5=0
cali6=0
cali=[]
for x in df['Overall']:
    if x < 50:
        cali1 =cali1 + 1
    elif x < 60:
        cali2 =cali2 + 1
    elif x < 70:
        cali3 =cali3 + 1
    elif x < 80:
        cali4 =cali4 + 1
    elif x < 90:
        cali5 =cali5 + 1
    else:
        cali6 =cali6 + 1

cali.append(cali1)
cali.append(cali2)
cali.append(cali3)
cali.append(cali4)
cali.append(cali5)
cali.append(cali6)
cali
```
